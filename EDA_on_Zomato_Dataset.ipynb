```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

Now as there are Country Code in both the datset we can combine the datset in the next block using pd.merge

```python
final_df= pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#Pie chart
plt.pie(country_val, labels=country_names)
```

```python
#pie chart - Top 3 Countries that uses Zomato
plt.pie(country_val[:3], labels = country_names[:3], autopct='%1.2f%%')
```

Observation : Zomato maximum records or transaction are from India After that  USA and then UK.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating Count'})
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating Count'})
```

```python
ratings
```

##Observation
When rating is between 4.5 to 4.9 ---> Excellent

When rating is between 4.0 to 3.4 ---> Very Good

When rating is between 3.5 to 3.9 ---> good

When rating is between 3.0 to 3.4 ---> average

When rating is between 2.5 to 2.9 ---> average

When rating is between 2.0 to 2.4 ---> poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette = ['white', 'red','orange', 'yellow', 'green','green'])
```

##Observation
1- Not Rated count is very High.

2-Max no. of ratings are between 2.5 to 3.4

```python
#Count plot
sns.countplot(x='Rating color', data=ratings, palette = ['white', 'red','orange', 'yellow', 'green','green'])
```
